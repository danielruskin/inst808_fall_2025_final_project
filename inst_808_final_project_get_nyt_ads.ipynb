```python
import pandas as pd
import os
from pdf2image import convert_from_path
import pytesseract
```

```python
data_path = '/content/drive/My Drive/Courses/INST808/nyt_ads'
ads_text = [] # create list of text for all ads, where each element represents the pages of one ad
for f in os.listdir(data_path): # go through each of the 76 pdfs
  pages = convert_from_path(os.path.join(data_path, f), dpi=300) # convert file to image
  text = [] # add one elemnt in list for each page in the pdf
  for i, image in enumerate(pages): # for every page...
    text.append(pytesseract.image_to_string(image)) # use pytesseract to extract text from image
    text.append(text)
  ads_text.append(text) # append to list of text for all ads
```

```python
results = pd.DataFrame({"text": ads_text})
results.head()
```

```python
# Save ads to file
file_path = os.path.join('/content/drive/My Drive/Courses/INST808/', 'nyt_ads.tsv')
results.to_csv(file_path, sep="\t", index=False)
```
